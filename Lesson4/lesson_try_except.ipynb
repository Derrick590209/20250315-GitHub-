```python
try:
  print(10/0)
except:
  print("程式結束")
```

```python
try:
  age = int(input("請輸入年紀:"))
  print(f"你輸入的年紀是{age}歲")
except:
  print(f"輸入的年紀錯誤")
```

```python
try:
  num = int(input("輸入數字"))
  print(num)

except Exception as a:
  print(f"錯誤內容:{a}")

print("程式結束")
```
