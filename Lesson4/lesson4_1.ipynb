```python

heigh = float(input("請輸入您的身高:"))
weigh = float(input("請輸入您的體重:"))
bmi =  weigh / (heigh/100)**2
print(f"您的身高為{heigh}公分,體重為{weigh}公斤,BMI為:{bmi:.2f}")
```

```python
# 學生總分為300
# 要檢查學生輸入的分數,是否是0~300
# 有些學生可以加分5%
# 學生分數加分後,不可以超過300

scores = int(input("請輸入學生分數(最高300分):"))
if scores >=0 and scores <=300:
  is_add = input("學生是否符合加分條件(y,n)")

  # 巢狀判斷 處理兩個條件
  if is_add == "y":
    scores *=1.05
    if scores > 300:
      scores = 300

  print(round(scores))
else:
  print("輸入的範圍,必需在0~300之間")

print("應用程式結束")
```

```python
try:
  raise NameError


except TypeError: #指定錯誤
  print("資料類型出錯了")
except ValueError: #指定錯誤
  print("值出錯了")
except:
  print("出錯了")

print("應用程式結束")
```

- 可以多個except程式區塊
- except程式區塊  只執行一個後  跳出try_except
- except(所有錯誤, 一定要放在最後面)

```python
try:
  raise NameError
except:
  print("出錯了")

except TypeError: #指定錯誤
  print("資料類型出錯了")
except ValueError: #指定錯誤
  print("值出錯了")


print("應用程式結束")
```
