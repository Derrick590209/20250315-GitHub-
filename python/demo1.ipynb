# 這是一個正弦波的圖表範例

```python
import numpy as np
import matplotlib.pyplot as plt

# Explicitly set sans-serif font to a commonly available one
# to prevent Matplotlib from searching for Noto Sans CJK JP when not needed.
plt.rcParams['font.sans-serif'] = ['DejaVu Sans', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = True # Ensures minus signs are displayed correctly with non-CJK fonts

# Generate x values in degrees from 0 to 360
x_degrees = np.linspace(0, 360, 400)

# Convert degrees to radians for the sine function
x_radians = np.deg2rad(x_degrees)

# Calculate the sine values
y = np.sin(x_radians)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(x_degrees, y, label='sin(x)', color='red')

# Add title and axis labels in English
plt.title('Sine Wave Example (0-360 Degrees)')
plt.xlabel('Angle (Degrees)')
plt.ylabel('sin(x) Value')

# Set x-axis ticks to common degree values for clarity
plt.xticks([0, 90, 180, 270, 360])

# Add grid lines
plt.grid(True)

# Add legend
plt.legend()

# Display the plot
plt.show()
```
